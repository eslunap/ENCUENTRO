**Se genera Lista para discriminar los paises de LATAM**

```python
country_codes_latam_a3 = [
    "ARG",  # Argentina
    "BOL",  # Bolivia
    "BRA",  # Brasil
    "CHL",  # Chile
    "COL",  # Colombia
    "CRI",  # Costa Rica
    "CUB",  # Cuba
    "ECU",  # Ecuador
    "SLV",  # El Salvador
    "GTM",  # Guatemala
    "HND",  # Honduras
    "MEX",  # México
    "NIC",  # Nicaragua
    "PAN",  # Panamá
    "PRY",  # Paraguay
    "PER",  # Perú
    "PRI", # Puerto Rico
    "DOM",  # República Dominicana
    "URY",  # Uruguay
    "VEN"   # Venezuela
]

print(country_codes_latam_a3)
```

**Para conocer el impacto de COVID en los diferentes años, se crea la lista con el tiempo de impacto**

```python
covid_year = [
    "2020",
    "2021",
    "2022",
    "2023",

]

print(covid_year)
```

# **Se saca la información de los paises de LATAM vs Años de covid, en donde hay una tendencia estable de PIB del sector en cada una de las naciones. No obstante, hay diferentes paises como Nicaragua y Bolivia que en dichos años fueron mas productivos en la agricultura, silvicultura y pesca**

```python
# Filtrar solo los países de LATAM usando 'Country Code'
df_latam = df2[df2['Country Code'].isin(country_codes_latam_a3)]
df_latam = df_latam.sort_values(by='Promedio', ascending=False)
df_latam
```

```python
import seaborn as sns
```

```python
# Calcular PIB total por país (sumando los años disponibles)
df_latam['PIB_Total'] = df_latam[covid_year].sum(axis=1)


# Ordenar los países de mayor a menor PIB total
df_latam = df_latam.sort_values(by='PIB_Total', ascending=False)
# Transformar los datos a formato largo (melt) para Seaborn
df_melted = df_latam.melt(id_vars=['Country Name'],
                          value_vars=covid_year,
                          var_name='Año',
                          value_name='PIB')


# Crear la figura
plt.figure(figsize=(12, 6))


# Graficar el histograma apilado
sns.histplot(data=df_melted, x='Country Name', hue='Año', weights='PIB', multiple='stack', palette='tab10', edgecolor='black')


# Graficar la curva de densidad para cada año
for year in covid_year:
    sns.kdeplot(df_melted[df_melted['Año'] == year]['PIB'], label=year, linewidth=2)


# Personalizar la gráfica
plt.xlabel('País')
plt.ylabel('PIB')
plt.title('Distribución del PIB en Latinoamérica (2020-2023) Impacto Covid')
plt.xticks(rotation=90)
plt.legend(title='Año')
plt.grid(axis='y', linestyle='--', alpha=1)


# Mostrar la gráfica
plt.show()
```

```python
# Seleccionar las columnas de los años de interés y calcular el promedio
df_latam['PIB_Promedio'] = df_latam[covid_year].mean(axis=1)


# Ordenar de mayor a menor
df_latam = df_latam.sort_values(by='PIB_Promedio', ascending=False)


# Crear la gráfica
plt.figure(figsize=(12, 6))
bars = plt.bar(df_latam['Country Name'], df_latam['PIB_Promedio'], color='skyblue')


# Agregar etiquetas de datos sobre cada barra
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, f'{yval:,.0f}',
             ha='center', va='bottom', fontsize=10, rotation=0)


# Personalizar la gráfica
plt.xlabel('País')
plt.ylabel('PIB Promedio (2020-2023)')
plt.title('Promedio del PIB en Latinoamérica durante la Pandemia')
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.7)
```

# **Se analiza la evolución de este sector en los países de LATAM desde su inicio hasta la actualidad, observando que países como Colombia y Brasil han incrementado su participación en la economía**

```python
# Variación porcentual del PIB (Crecimiento/Decrecimiento)
df_latam['Variacion_PIB'] = ((df_latam['2023'] - df_latam['2020']) / df_latam['2020']) * 100
df_latam_sorted = df_latam.sort_values(by='Variacion_PIB', ascending=False)


plt.figure(figsize=(10, 6))
sns.barplot(y=df_latam_sorted['Country Name'], x=df_latam_sorted['Variacion_PIB'], palette='coolwarm')
plt.xlabel('Variación % del PIB (2020-2023)')
plt.ylabel('País')
plt.title('Crecimiento o Decrecimiento del PIB en LATAM (2020-2023)')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()

```
