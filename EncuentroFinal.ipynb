**Se genera Lista para discriminar los paises de LATAM**

```python
country_codes_latam_a3 = [
    "ARG",  # Argentina
    "BOL",  # Bolivia
    "BRA",  # Brasil
    "CHL",  # Chile
    "COL",  # Colombia
    "CRI",  # Costa Rica
    "CUB",  # Cuba
    "ECU",  # Ecuador
    "SLV",  # El Salvador
    "GTM",  # Guatemala
    "HND",  # Honduras
    "MEX",  # M√©xico
    "NIC",  # Nicaragua
    "PAN",  # Panam√°
    "PRY",  # Paraguay
    "PER",  # Per√∫
    "PRI", # Puerto Rico
    "DOM",  # Rep√∫blica Dominicana
    "URY",  # Uruguay
    "VEN"   # Venezuela
]

print(country_codes_latam_a3)
```

**Para conocer el impacto de COVID en los diferentes a√±os, se crea la lista con el tiempo de impacto**

```python
covid_year = [
    "2020",
    "2021",
    "2022",
    "2023",

]

print(covid_year)
```

# **Se saca la informaci√≥n de los paises de LATAM vs A√±os de covid, en donde hay una tendencia estable de PIB del sector en cada una de las naciones. No obstante, hay diferentes paises como Nicaragua y Bolivia que en dichos a√±os fueron mas productivos en la agricultura, silvicultura y pesca**

```python
# Filtrar solo los pa√≠ses de LATAM usando 'Country Code'
df_latam = df2[df2['Country Code'].isin(country_codes_latam_a3)]
df_latam = df_latam.sort_values(by='Promedio', ascending=False)
df_latam
```

```python
import seaborn as sns
```

```python
# Calcular PIB total por pa√≠s (sumando los a√±os disponibles)
df_latam['PIB_Total'] = df_latam[covid_year].sum(axis=1)


# Ordenar los pa√≠ses de mayor a menor PIB total
df_latam = df_latam.sort_values(by='PIB_Total', ascending=False)
# Transformar los datos a formato largo (melt) para Seaborn
df_melted = df_latam.melt(id_vars=['Country Name'],
                          value_vars=covid_year,
                          var_name='A√±o',
                          value_name='PIB')


# Crear la figura
plt.figure(figsize=(12, 6))


# Graficar el histograma apilado
sns.histplot(data=df_melted, x='Country Name', hue='A√±o', weights='PIB', multiple='stack', palette='tab10', edgecolor='black')


# Graficar la curva de densidad para cada a√±o
for year in covid_year:
    sns.kdeplot(df_melted[df_melted['A√±o'] == year]['PIB'], label=year, linewidth=2)


# Personalizar la gr√°fica
plt.xlabel('Pa√≠s')
plt.ylabel('PIB')
plt.title('Distribuci√≥n del PIB en Latinoam√©rica (2020-2023) Impacto Covid')
plt.xticks(rotation=90)
plt.legend(title='A√±o')
plt.grid(axis='y', linestyle='--', alpha=1)


# Mostrar la gr√°fica
plt.show()
```

```python
# Seleccionar las columnas de los a√±os de inter√©s y calcular el promedio
df_latam['PIB_Promedio'] = df_latam[covid_year].mean(axis=1)


# Ordenar de mayor a menor
df_latam = df_latam.sort_values(by='PIB_Promedio', ascending=False)


# Crear la gr√°fica
plt.figure(figsize=(12, 6))
bars = plt.bar(df_latam['Country Name'], df_latam['PIB_Promedio'], color='skyblue')


# Agregar etiquetas de datos sobre cada barra
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, f'{yval:,.0f}',
             ha='center', va='bottom', fontsize=10, rotation=0)


# Personalizar la gr√°fica
plt.xlabel('Pa√≠s')
plt.ylabel('PIB Promedio (2020-2023)')
plt.title('Promedio del PIB en Latinoam√©rica durante la Pandemia')
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.7)
```

# **Se analiza la evoluci√≥n de este sector en los pa√≠ses de LATAM desde su inicio hasta la actualidad, observando que pa√≠ses como Colombia y Brasil han incrementado su participaci√≥n en la econom√≠a**

```python
# Variaci√≥n porcentual del PIB (Crecimiento/Decrecimiento)
df_latam['Variacion_PIB'] = ((df_latam['2023'] - df_latam['2020']) / df_latam['2020']) * 100
df_latam_sorted = df_latam.sort_values(by='Variacion_PIB', ascending=False)


plt.figure(figsize=(10, 6))
sns.barplot(y=df_latam_sorted['Country Name'], x=df_latam_sorted['Variacion_PIB'], palette='coolwarm')
plt.xlabel('Variaci√≥n % del PIB (2020-2023)')
plt.ylabel('Pa√≠s')
plt.title('Crecimiento o Decrecimiento del PIB en LATAM (2020-2023)')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()

```
